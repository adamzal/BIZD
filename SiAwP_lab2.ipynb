```python
import numpy as np
from scipy.stats import describe, bernoulli, binom, poisson, norm
import matplotlib.pyplot as plt
```

# Zadanie 1

```python
wartosci = [1, 2, 3, 4, 5, 6]
prawdopodobienstwa = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
```

```python
mean = sum(wartosci[i] * prawdopodobienstwa[i] for i in range(len(wartosci)))
mean
```

```python
variance = sum((wartosci[i] - mean)**2 * prawdopodobienstwa[i] for i in range(len(wartosci)))
variance
```

```python
stdev = variance**0.5
stdev
```

# Zadanie 2

```python
p_ber = 0.3
n = 100
```

```python
proby_ber = np.random.binomial(1, p_ber, n)
proby_ber
```

```python
n_dwum = 10
p_dwum = 0.5
```

```python
proby_dwum = np.random.binomial(n_dwum, p_dwum, n)
proby_dwum
```

```python
lambda_poisson = 2.5
```

```python
proby_poisson = np.random.poisson(lambda_poisson, n)
proby_poisson
```

# Zadanie 4

```python
x_ber = np.array([0, 1])
pmf_ber = bernoulli.pmf(x_ber, p_ber)
```

```python
x_dwum = np.arange(0, n_dwum+1)
pmf_dwum = binom.pmf(x_dwum, n_dwum, p_dwum)
```

```python
x_poisson = np.arange(0, 10)
pmf_poisson = poisson.pmf(x_poisson, lambda_poisson)
```

```python
plt.bar(x_ber, pmf_ber, width=0.4)
plt.title('Rozkład Bernoulliego')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')
plt.show()
```

```python
plt.bar(x_dwum, pmf_dwum, width=0.4)
plt.title('Rozkład Dwumianowy')
plt.xlabel('Liczba sukcesów')
plt.ylabel('Prawdopodobieństwo')
plt.show()
```

```python
plt.bar(x_poisson, pmf_poisson, width=0.4)
plt.title('Rozkład Poissona')
plt.xlabel('Liczba zdarzeń')
plt.ylabel('Prawdopodobieństwo')
plt.show()
```

# Zadanie 5

```python
print(f'Rozkład prawdopodobieństwa dla rozkładu Dwumianowego (n=20, p=0.4):')
for k, prawdopodobienstwo in zip(x_dwum, pmf_dwum):
    print(f'P(X={k}) = {prawdopodobienstwo}')
```

```python
suma_prawdopodobienstw = np.sum(pmf_dwum)
print(f'\nSuma prawdopodobieństw: {suma_prawdopodobienstw}')
```

```python
if np.isclose(suma_prawdopodobienstw, 1.0):
    print('Suma prawdopodobieństw jest równa 1.')
else:
    print('Uwaga: Suma prawdopodobieństw nie jest równa 1.')
```

# Zadanie 7

```python
dane1 = np.random.normal(1, 2, 1000)
dane_standardowe = np.random.standard_normal(1000)
dane2 = np.random.normal(-1, .5, 1000)
```

```python
plt.hist(dane1, bins=30, density=True, alpha=0.7, color='blue', label='N(1, 2)')
plt.hist(dane_standardowe, bins=30, density=True, alpha=0.7, color='orange', label='Standardowy normalny')

x = np.linspace(-5, 3, 100)
y = norm.pdf(x, loc=-1, scale=.5)

plt.plot(x, y, 'r-', label='N(-1, 0.5)')
plt.legend()
plt.xlabel('Wartość')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title('Histogramy i wykresy gęstości dla różnych rozkładów normalnych')
plt.show()
```
